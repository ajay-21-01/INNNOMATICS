```python
# Problem 1 - Unique words in a sentence

sentence = input("Enter sentence: ")

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2 - Highest salary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
# Problem 3 - Max and Min in list

numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4 - Count products above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem 5 - Attendance percentage

attendance = ["P", "P", "A", "P", "P"]

present = attendance.count("P")
total = len(attendance)

percentage = (present / total) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6 - Remove duplicate phone numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7 - Character frequency

text = "python"

frequency = {}

for ch in text:
    if ch in frequency:
        frequency[ch] += 1
    else:
        frequency[ch] = 1

print(frequency)
```

```python
# Problem 8 - Convert list to tuple

numbers = [10, 20, 30]

numbers_tuple = tuple(numbers)

print("Tuple:", numbers_tuple)
```

```python
# Problem 9 - Check key in dictionary

employees = {
    "Ravi": 75000,
    "Anita": 68000
}

key = "Ravi"

if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10 - Average marks

marks = [70, 80, 65, 75]

average = sum(marks) / len(marks)

print("Average Marks:", average)
```
