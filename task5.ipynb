```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):

    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    # Conditional check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


likes = list(map(int,input().split()))
total, status = analyze_post_engagement(likes)

print("Total Likes:", total)
print("Post Status:", status)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
    """
    Checks stock level and returns status.
    """
    if stock < 10:
        return "Low Stock Alert"
    else:
        return "Stock Sufficient"



stock = 6
print("Medicine Stock:", stock)
print("Status:", check_medicine_stock(stock))
```

```python
# Problem 3: Rainfall Adequacy Checker

def check_rainfall(rainfall_data, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


rainfall = [70, 75, 68, 80]
average, status = check_rainfall(rainfall, 70)

print("Average Rainfall:", average)
print("Rainfall Status:", status)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    """
    Detects duplicate usernames using set.
    """
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"



users = ["ajay", "rahul", "ajay", "sita"]
print("Duplicate Accounts Found:", detect_duplicate_accounts(users))
```

```python
# Problem 5: Appointment Eligibility Checker

def check_appointment_eligibility(age):
    """
    Checks if patient is eligible for appointment.
    """
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"


age = 21
print("Patient Age:", age)
print("Eligibility Status:", check_appointment_eligibility(age))
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    """
    Filters crop prices above ₹2000.
    """
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
print("Premium Crops:", filter_premium_crops(crop_prices))
```

```python
# Problem 7: Application Health Checker

def check_application_health(error_count):
    """
    Evaluates system health based on error count.
    """
    if error_count == 0:
        return "Healthy"
    elif error_count <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"


# Example usage
errors = 7
print("Error Count:", errors)
print("System Status:", check_application_health(errors))
```

```python
# Problem 8: Daily Transaction Limit Checker

def check_transaction_limit(amount):
    """
    Validates daily transaction limit.
    """
    daily_limit = 50000

    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"


# Example usage
amount = 60000
print("Transaction Amount:", amount)
print("Transaction Status:", check_transaction_limit(amount))
```

```python
# Problem 9: Attendance Eligibility System

def check_attendance(attendance_list):
    """
    Calculates attendance percentage and checks eligibility.
    """
    total_days = len(attendance_list)
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return percentage, status


# Example usage
attendance = ["P", "P", "A", "P", "P"]
percentage, status = check_attendance(attendance)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", status)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    """
    Calculates electricity bill based on slab system
    and classifies usage level.
    """
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units = 250
bill, status = calculate_electricity_bill(units)

print("Total Bill:", bill)
print("Usage Classification:", status)
```
